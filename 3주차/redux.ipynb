```python
# 리덕스는 데이터를 한군데에 모아놓고, 꺼내볼수있게해주는(참조) 패키지
#리덕의 주요 용어
1.State = 리덕스에서의 데이터 --- 딕셔너리 형태 ({[key]: value})
2.Action = 데이터 상태를 변경하고 싶을때 발생 (수정할때)  --- {type: 'CHANGE_STATE', data: {...}}
3.ActionCreator = Action을 위한 함수 액션을 만들기 위해 사용한다.
    const changeState = (new_data) => { # Action을 return해준다.
        return {
            type: 'CHANGE_STATE',
            data: new_data
        }
    }
4.Reducer = 저장된 데이터의 상태를 변경해주는 함수
const initialState = {
    name: 'mean0'
}
function reducer(state = initialState, action) {
    switch(action.type){
        // action의 타입마다 케이스문을 걸어주면, 
        // 액션에 따라서 새로운 값을 돌려줍니다!
        case CHANGE_STATE: 
            return {name: 'mean1'};

        default: 
            return false;
    }
}

5.Store = 데이터를 볼수있게 해준다.(리덕스를 적용하기 위함) --- 딕셔너리 형태 ({[key]: value})  json처럼 생김

6.dispatch(중요) = Action을 발생시키는 역할 action return를 해준다.


ActionCreator(액션함수)만들고 -> Reducer가 action의 객체를 받아서 -> 데이터 수정 -> Return
```

```python
# Redux 3가지 특징
1. Store는 딱 1개만 쓴다. (한프로젝트에 1개만쓴다.)
2. Store의 state(데이터)는 오직 action으로만 변경할 수 있다
 *** Store 안에 있는 데이터는 *무조건* Action을 발생시켜 데이터 수정을 이루어야함!! ***
3. Reducer는 어떤 요청이 와도 리듀서는 같은 동작을 해야한다!
   **리듀서는 순수한 함수여야 한다는 말입니다.
        파라미터 외의 값에 의존하지 않아야함.
        파라미터가 같으면,항상 같은 값을 반환
```

```python
상태관리 흐름

1. Store를 Component들에 연결시켜준다.
2. Component는 데이터 상태를 바꿔달라는 호출 즉 Action를 부른다.
3. Reducer는 호출에 응답하여 데이터를 새로운 값으로 만든다.(수정) <--> Reducer 함수실행
4. 바뀐 새로운 상태값을 Store에 저장
5. Component는 새로운 값을 받아온다(props로 받아옴) -- props를 받아옴으로써 리랜더링됨
```
